# Etude Bi-varié

# Etude uni-varié

Notes en fonction de la volonté de poursuivre ses études (higher)

```python
sns.violinplot("G3", "higher", data=df,
               palette=["lightblue", "lightgreen"]);
```

Boxplots des notes obtenues en G3 en fonction du nombre de redoublements

```python
sns.boxplot(x='G3' ,y='failures', data=df)
```

```python
with sns.axes_style('white'):
    g = sns.jointplot("G3", "Medu", df, kind='hex')
```

```python
sns.boxplot(x='G3' ,y='Mjob', data=df)
```
